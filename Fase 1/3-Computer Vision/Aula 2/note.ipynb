```python
from paddleocr import PaddleOCR
import re

def extract_text_from_pdf(pdf_path):
    # Inicializa o OCR
    ocr = PaddleOCR(use_angle_cls=True, lang='en')
    
    # Realiza a extração de texto
    result = ""
    for line in ocr.ocr(pdf_path):
        for word in line:
            result += word[1][0] + " "
        result += "\n"
    
    return result

# Caminho do arquivo PDF ou imagem
pdf_path = "Release de Resultados do 1T24.pdf"

# Extrai o texto do PDF/imagem
ocr_text = extract_text_from_pdf(pdf_path)

# Imprime o texto extraído
print("Texto Extraído:")
print(ocr_text)




# Definições de padrões (Expressões Regulares)
padrao_cpf_cnpj = r'[0-9]{3}\.[0-9]{3}\.[0-9]{3}-[0-9]{2}|[0-9]{2}\.[0-9]{3}\.[0-9]{3}/[0-9]{4}-[0-9]{2}'
padrao_nome_cliente = r'Nome do Cliente:\s*(.*)'
padrao_num_cliente = r'Número do Cliente:\s*(\d+)'
padrao_data_leitura = r'Data de Leitura:\s*(\d{2}/\d{2}/\d{4})'
padrao_data_vencimento = r'Data de Vencimento:\s*(\d{2}/\d{2}/\d{4})'
padrao_valor_total = r'Valor Total:\s*R\$\s*([\d,]+.\d{2})'
padrao_descricao_faturamento = r'Descrição do Faturamento:\s*(.*)'
padrao_referencia_faturamento = r'Referência de Faturamento:\s*(\d{2}/\d{2}/\d{4})'
padrao_qr_code = r'QRCode para pagamento via PIX:\s*(.*)'


def extrair_informacoes(texto, padrao):
    match = re.search(padrao, texto)
    if match:
        return match.group(1)
    else:
        return None

# Extração de Informações
ocr_text = "..."  # Texto extraído pelo OCR
cpf_cnpj = extrair_informacoes(ocr_text, r'CPF/CNPJ:\s*(\d+)')
nome_cliente = extrair_informacoes(ocr_text, r'Nome:\s*(\w+\s\w+)')
num_cliente = extrair_informacoes(ocr_text, r'Número do Cliente:\s*(\d+)')
data_vencimento = extrair_informacoes(ocr_text, r'Data de Vencimento:\s*(\d{2}/\d{2}/\d{4})')
valor_total = extrair_informacoes(ocr_text, r'Valor Total:\s*R\$\s*(\d+,\d{2})')
referencia_faturamento = extrair_informacoes(ocr_text, r'Referência de Faturamento:\s*(\d+)')

# Verificação de Presença de QR Code
padrao_qr_code = r'QRCode:\s*(\w+)'
qr_code_presente = bool(re.search(padrao_qr_code, ocr_text))

# Exibição das Informações Extraídas
print(f"CPF/CNPJ: {cpf_cnpj}")
print(f"Nome do Cliente: {nome_cliente}")
print(f"Número do Cliente: {num_cliente}")
print(f"Data de Vencimento: {data_vencimento}")
print(f"Valor Total: {valor_total}")
print(f"Referência de Faturamento: {referencia_faturamento}")
print(f"QR Code Presente: {qr_code_presente}")
```
